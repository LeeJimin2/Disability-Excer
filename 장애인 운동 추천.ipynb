```python
path = "C:/Users/main/Desktop/데이터경진대회/체력측정데이터/"
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')]
```

```python
# 체력측정데이터
# =============================================================================
mse_df = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(path + i)
    mse_df = pd.concat([mse_df,data])


mse_df = mse_df.reset_index(drop = True)
```

```python
# =============================================================================
# 운동처방데이터
# =============================================================================
path = "C:/Users/main/Desktop/데이터경진대회/운동처방/"
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')] 

prs_df = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(path + i)
    prs_df = pd.concat([prs_df,data])


prs_df = prs_df.reset_index(drop = True)
```

```python
# =============================================================================
# 나이 범주화
# =============================================================================
def age_categorize(age):
    age = (age//10) * 10
    return age
```
